## Part 2

* The `mount_everest_deaths` dataset has its own index column provided in the dataset.  When importing it, use the existing index column.
* Which columns have zero null values?
* Which column has the most null values?


```python
deaths = load_dataset('mount_everest_deaths.csv', index_col=0)
```

```python
deaths
```

```python
deaths.info()
```

```python
zero_nulls = deaths.columns[deaths.isnull().sum()==0]
print(f"The Columns with Zero Null Values: {zero_nulls}")
```

```python
null_counts = deaths.isnull().sum()
maximum = null_counts.max()
max_nulls = null_counts[null_counts == maximum].index
print(f"The Columns with Maximum Null Values: {list(max_nulls)} ({maximum} nulls)")
```
