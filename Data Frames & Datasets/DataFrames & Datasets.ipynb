
# **King County Home Sales Data Set**


This dataset includes the home sales from 2014-2015 in King County, WA (the county Seattle is located in)

## Description

This dataset includes the home sales from 2014–2015 in King County, WA (the county Seattle is located in):

1. **id** – House's unique ID  
2. **date** – Sale date  
3. **price** – Sale price  
4. **bedrooms** – Number of bedrooms  
5. **bathrooms** – Number of bathrooms  
6. **sqft_living** – Living space square footage  
7. **sqft_lot** – Total lot square footage  
8. **floors** – Number of floors  
9. **waterfront** – Is the house waterfront (1) or not (0)  
10. **view** – Rating from 0 to 4 of how good the view from the house is  
11. **condition** – Rating from 1 (poor) to 5 (very good) of the house's condition  
12. **grade** – Rating from 1 to 13 representing construction quality  
13. **sqft_above** – Square footage of the interior above ground level  
14. **sqft_basement** – Square footage of the interior below ground level  
15. **yr_built** – Year the house was initially built  
16. **yr_renovated** – Year of the house’s last renovation (if any)  
17. **zipcode** – Zipcode where the house is located  
18. **lat** – Property’s latitude  
19. **long** – Property’s longitude  
20. **sqft_living15** – Average interior space square footage of the nearest 15 neighbors  
21. **sqft_lot15** – Average lot square footage of the nearest 15 neighbors

## read_csv with non-comma separators
The `netflix_titles.csv` file is actually pipe (`|`) separated.  We can tell `read_csv` to expect this using the `sep="|"` option.

Additionally, the netflix dataset includes an index column as the very first column.  We can let `read_csv` know using `index_col=0`

```python

netflix = load_dataset("netflix_titles.csv", sep="|",index_col=0)
netflix
```

```python
netflix = load_dataset("netflix_titles.csv", sep="|",index_col=3)
netflix
```
